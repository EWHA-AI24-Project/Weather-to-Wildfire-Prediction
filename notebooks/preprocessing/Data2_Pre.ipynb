```python
from google.colab import drive
drive.mount('/content/drive')

file_path = '/content/dataset_2_ordered.csv'
import pandas as pd
df = pd.read_csv(file_path)
df.head()
```

```python
# PRECIPITATION: inch -> mm
df['PRECIPITATION'] = df['PRECIPITATION'] * 25.4

# TEMPERATUER: °F -> °C
df['MAX_TEMP'] = (df['MAX_TEMP'] - 32) * 5/9
df['MIN_TEMP'] = (df['MIN_TEMP'] - 32) * 5/9

# WINDSPEED: mph -> m/s
df['AVG_WIND_SPEED'] = df['AVG_WIND_SPEED'] * 0.44704
```

```python
df['TEMP_RANGE'] = df['MAX_TEMP'] - df['MIN_TEMP']
```

```python
df_encoded = pd.get_dummies(df, columns=['SEASON'], prefix='SEASON')
```

```python
SEASON_cols = [col for col in df_encoded.columns if col.startswith('SEASON_')]
df_encoded[SEASON_cols] = df_encoded[SEASON_cols].astype(int)
```
