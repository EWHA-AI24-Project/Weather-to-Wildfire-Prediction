```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('./dataset_2.csv')
df
```

```python
# 결측치 개수와 비율 확인
missing_info = df.isnull().sum().to_frame(name='MissingCount')
missing_info['Total'] = len(df)
missing_info['MissingRatio (%)'] = (missing_info['MissingCount'] / missing_info['Total']) * 100
missing_info = missing_info.drop(columns='Total')

# 결과 보기
print(missing_info)
```
